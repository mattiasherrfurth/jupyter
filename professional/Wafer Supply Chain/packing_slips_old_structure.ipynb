```python
import os
import pyodbc
```

```python
def postgresProdDsn(query):
    cnxn = pyodbc.connect("DSN=PostgresBronxProd")
    cursor = cnxn.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    cnxn.close()
    return data

def postgresSboxDsn(query):
    cnxn = pyodbc.connect("DSN=PostgresBronxSandbox")
    cursor = cnxn.cursor()
    cursor.execute(query)
    data = cursor.fetchall()
    cnxn.close()
    return data
```

```python
query = '''
select distinct f.filename
from bronx.tek_pack_slip tps
join bronx.files_metadata f on tps.file_id = f.file_id
'''
data = postgresSboxDsn(query)
tblfiles = [x[0] for x in data]
```

```python
wps_dir = r'C:\path\to\folder\WaferPackingSlips'
mps_dir = r'C:\path\to\folder\MCMPackingSlips'
```

```python
unloaded = []
for n in os.listdir(wps_dir):
    with open(wps_dir + '\\' + n) as f:
        lines = f.readlines()
        if lines[1].startswith('<!-- Generated by Oracle Reports'):
            print(n)
            unloaded += [n]

for n in os.listdir(mps_dir):
    with open(mps_dir + '\\' + n) as f:
        try:
            lines = f.readlines()
            if lines[1].startswith('<!-- Generated by Oracle Reports'):
                print(n)
                unloaded += [n]
        except UnicodeDecodeError as e:
            pass
```

```python
to_be_loaded = list(set(unloaded).difference(tblfiles))
for x in to_be_loaded:
    print(x)
```
